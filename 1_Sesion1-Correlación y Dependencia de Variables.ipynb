### üìå **Diferencia entre Correlaci√≥n y Causalidad**  
üìñ **Definici√≥n:**  
La **correlaci√≥n** mide el grado de relaci√≥n entre dos variables, mientras que la **causalidad** implica que una variable es responsable de los cambios en otra.  

üîπ **Ejemplo ilustrativo:**  
- Se observa una alta correlaci√≥n entre el consumo de helado y el n√∫mero de ataques de tibur√≥n.  
- ¬øSignifica esto que comer helado provoca ataques de tibur√≥n? ‚ùå **No**.  
- La verdadera causa es el verano: cuando hace calor, m√°s personas compran helado y m√°s personas nadan en el mar, aumentando la probabilidad de ataques.

### üìå **Tipos de Correlaci√≥n**  
üîπ **Correlaci√≥n positiva (+)** ‚Üí Aumenta una variable, aumenta la otra.  
üîπ **Correlaci√≥n negativa (-)** ‚Üí Aumenta una variable, disminuye la otra.  
üîπ **Correlaci√≥n nula (‚âà0)** ‚Üí No hay relaci√≥n clara entre las variables.  

üìä **Ejemplos gr√°ficos:**  
1. **Positiva**: Ingreso mensual vs. gasto en entretenimiento.  
2. **Negativa**: N√∫mero de horas de ejercicio vs. nivel de colesterol.  
3. **Nula**: Tama√±o del zapato vs. calificaciones acad√©micas.

### üìå **Coeficiente de Correlaci√≥n de Pearson**  
üìñ **Definici√≥n matem√°tica:**  
$
r = \frac{\sum (X_i - \bar{X}) (Y_i - \bar{Y})}{\sqrt{\sum (X_i - \bar{X})^2} \sqrt{\sum (Y_i - \bar{Y})^2}}
$

üìä **Interpretaci√≥n de valores:**  
- **r = 1** ‚Üí Correlaci√≥n perfectamente positiva.  
- **r = -1** ‚Üí Correlaci√≥n perfectamente negativa.  
- **r = 0** ‚Üí No hay relaci√≥n lineal.  

üìå **Limitaciones:**  
- No detecta relaciones no lineales.  
- Es sensible a valores at√≠picos.  

üìä **Ejemplo gr√°fico:**  

-----

```python
import pandas as pd
import seaborn as sns
df = pd.DataFrame({'A': [1, 2, 3, 4, 5], 
                   'B': [2, 4, 6, 8, 10], 
                   'C': [5, 3, 8, 2, 7]})

correlation_matrix = df.corr()
print(correlation_matrix)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos
np.random.seed(0)
x = np.linspace(0, 10, 100)

# Correlaci√≥n positiva
y_positive = 2 * x + np.random.normal(0, 2, len(x))

# Correlaci√≥n negativa
y_negative = -2 * x + np.random.normal(0, 2, len(x))

# Correlaci√≥n cercana a cero (distribuci√≥n aleatoria)
y_zero = np.random.normal(5, 2, len(x))

# Crear subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Gr√°fico de correlaci√≥n positiva
axes[0].scatter(x, y_positive, color='green')
axes[0].set_title("Correlaci√≥n Positiva")
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")

# Gr√°fico de correlaci√≥n negativa
axes[1].scatter(x, y_negative, color='red')
axes[1].set_title("Correlaci√≥n Negativa")
axes[1].set_xlabel("X")
axes[1].set_ylabel("Y")

# Gr√°fico de correlaci√≥n cero
axes[2].scatter(x, y_zero, color='blue')
axes[2].set_title("Correlaci√≥n Cero")
axes[2].set_xlabel("X")
axes[2].set_ylabel("Y")

plt.tight_layout()
plt.show()
```

```python
def classify_correlation(r):
    """
    Clasifica la correlaci√≥n en positiva, negativa o nula
    basado en el valor del coeficiente de Pearson (r)
    """
    if r > 0.1:
        return "Positiva"
    elif r < -0.1:
        return "Negativa"
    else:
        return "Nula o muy d√©bil"
```

```python
import pandas as pd
# Import required libraries
# matplotlib.pyplot: Used for creating the visualization and controlling figure properties
import matplotlib.pyplot as plt

df = pd.read_csv('data/score_updated.csv')
df.shape
```

```python
# Create single scatter plot
fig, ax = plt.subplots(figsize=(10, 8))
fig.suptitle('Hours vs Scores - Dispersion Analysis', fontsize=16, fontweight='bold')

correlation = df['Hours'].corr(df['Scores'])

# Create scatter plot
ax.scatter(df['Hours'], df['Scores'], alpha=0.7, color='blue', s=60)
ax.set_xlabel('Hours Studied')
ax.set_ylabel('Test Scores')
ax.set_title(f'Correlaci√≥n: {correlation:.2f} / Con Correlacion {classify_correlation(correlation)}', fontsize=14)
ax.grid(True, alpha=0.3)
```

**Ice Cream Selling dataset**: El conjunto de datos muestra la relaci√≥n entre la temperatura y las ventas de helados. Sirve como un ejemplo pr√°ctico para comprender e implementar la regresi√≥n polin√≥mica, una t√©cnica poderosa para modelar relaciones no lineales en los datos.

üìå **Preguntas:**
- ¬øC√≥mo cambia la correlaci√≥n en distintas estaciones?
- ¬øExisten valores at√≠picos? ¬øC√≥mo afectan la interpretaci√≥n?
- ¬øQu√© otras variables podr√≠an influir en la venta de helado?

```python
# Gr√°fica de dispersi√≥n y an√°lisis de correlaci√≥n
#!pip install scipy
#!pip install scikit-learn

from scipy.stats import pearsonr  # Para calcular correlaci√≥n lineal
from sklearn.metrics import r2_score  # Para calcular R^2 en regresi√≥n

# Crear gr√°fico de dispersi√≥n de temperatura vs ventas
plt.scatter(df['Temperature (¬∞C)'], df['Ice Cream Sales (units)'])
plt.xlabel('Temperature (¬∞C)')  # Etiqueta eje X
plt.ylabel('Ice Cream Sales (units)')  # Etiqueta eje Y
plt.title('Ventas de helado vs Temperatura')  # T√≠tulo del gr√°fico
plt.show()

# Calcular y mostrar correlaci√≥n lineal entre temperatura y ventas
r, _ = pearsonr(df['Temperature (¬∞C)'], df['Ice Cream Sales (units)'])
print(f"Correlaci√≥n lineal r: {r:.3f}")

# Realizar regresi√≥n polin√≥mica de grado 2
# Ajustar coeficientes del polinomio
coeffs = np.polyfit(df['Temperature (¬∞C)'], df['Ice Cream Sales (units)'], deg=2)
# Predecir valores usando el modelo polin√≥mico
sales_pred = np.polyval(coeffs, df['Temperature (¬∞C)'])
# Calcular R^2 para medir bondad de ajuste
r2 = r2_score(df['Ice Cream Sales (units)'], sales_pred)
print(f"Coeficiente de determinaci√≥n (R^2) polin√≥mico: {r2:.3f}")

# Detecci√≥n de valores at√≠picos usando m√©todo del rango intercuart√≠lico (IQR)
q1 = df['Ice Cream Sales (units)'].quantile(0.25)  # Primer cuartil
q3 = df['Ice Cream Sales (units)'].quantile(0.75)  # Tercer cuartil
iqr = q3 - q1  # Rango intercuart√≠lico
# Identificar outliers usando 1.5 * IQR como l√≠mite
outliers = df[(df['Ice Cream Sales (units)'] < (q1 - 1.5 * iqr)) |
              (df['Ice Cream Sales (units)'] > (q3 + 1.5 * iqr))]
print(f"N√∫mero de outliers detectados: {outliers.shape[0]}")
```
