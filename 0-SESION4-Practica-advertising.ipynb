## 2\. Supuestos Te贸ricos del Modelo de Regresi贸n Lineal 

Un modelo de regresi贸n lineal (conocido como OLS - *Ordinary Least Squares*) es confiable solo si se cumplen ciertos supuestos sobre los datos y los errores del modelo (residuos). Los m谩s importantes son (acr贸nimo **L.I.N.E.**):

1.  **Linealidad (L):** La relaci贸n entre las variables predictoras y la variable objetivo debe ser lineal. (Ya lo exploramos visualmente en el paso 1).
2.  **Independencia (I):** Los residuos (los errores de predicci贸n) deben ser independientes entre s铆. Es decir, el error de una observaci贸n no debe influir en el error de otra.
3.  **Normalidad (N):** Los residuos deben seguir una distribuci贸n normal. Esto es importante para las pruebas de hip贸tesis sobre los coeficientes.
4.  **Homocedasticidad (E - Equal Variance):** La varianza de los residuos debe ser constante a lo largo de todos los niveles de las variables predictoras. En otras palabras, la dispersi贸n de los errores debe ser la misma para valores peque帽os y grandes de los predictores.

No podemos verificar todos estos supuestos a煤n, pero los tendremos en mente y **volveremos a ellos** despu茅s de construir nuestro primer modelo.

## 3\. Regresi贸n Lineal Simple 

Comencemos con el caso m谩s sencillo: predecir las ventas usando **solo una** variable predictora. Bas谩ndonos en nuestro an谩lisis, `TV` es el mejor candidato.

**Modelo a estimar:** $Sales = \theta_0 + \theta_1 \times TV$

```python
# 1. Definir nuestras variables X e y
X = data[['TV']]
y = data['Sales']

# 2. Dividir los datos en conjuntos de entrenamiento y prueba
# Esto es CRUCIAL para evaluar si nuestro modelo generaliza bien a datos nuevos.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Crear y entrenar el modelo
model_simple = LinearRegression()
model_simple.fit(X_train, y_train)

# 4. Ver los par谩metros estimados
print(f"Ordenada al origen (胃): {model_simple.intercept_:.4f}")
print(f"Coeficiente para TV (胃): {model_simple.coef_[0]:.4f}")
```


**Visualicemos el modelo ajustado:**


```python
# 5. Visualizar la l铆nea de regresi贸n sobre los datos de prueba
plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color='blue', label='Datos Reales')
plt.plot(X_test, model_simple.predict(X_test), color='red', linewidth=2, label='L铆nea de Regresi贸n')
plt.title('Regresi贸n Lineal Simple: Ventas vs. Gasto en TV')
plt.xlabel('Gasto en TV (miles de USD)')
plt.ylabel('Ventas (miles de unidades)')
plt.legend()
plt.grid(True)
plt.show()
```


**Ahora, verifiquemos los supuestos (Normalidad y Homocedasticidad de los residuos):**


```python
# 6. Calcular los residuos en el conjunto de prueba
residuos = y_test - model_simple.predict(X_test)

# Gr谩fico de Normalidad (Histograma)
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.histplot(residuos, kde=True)
plt.title('Histograma de Residuos')

# Gr谩fico de Homocedasticidad
plt.subplot(1, 2, 2)
plt.scatter(model_simple.predict(X_test), residuos)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Predicciones vs. Residuos')
plt.xlabel('Valores Predichos')
plt.ylabel('Residuos')
plt.show()
```

## 4\. Regresi贸n Lineal M煤ltiple 锔

驴Y si usamos toda la informaci贸n que tenemos? Un modelo de regresi贸n m煤ltiple nos permite usar `TV`, `Radio` y `Newspaper` juntos para predecir las ventas.

**Modelo a estimar:** $Sales = \theta_0 + \theta_1 \times TV + \theta_2 \times Radio + \theta_3 \times Newspaper$


```python
# 1. Definir X con todas las variables predictoras
X_multi = data[['TV', 'Radio', 'Newspaper']]
y_multi = data['Sales']

# 2. Dividir los datos
X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(X_multi, y_multi, test_size=0.3, random_state=42)

# 3. Crear y entrenar el modelo
model_multiple = LinearRegression()
model_multiple.fit(X_train_m, y_train_m)

# 4. Ver los par谩metros
print("Ordenada al origen (胃):", model_multiple.intercept_)
print("\nCoeficientes (胃, 胃, 胃):")
print(pd.DataFrame(model_multiple.coef_, X_multi.columns, columns=['Coeficiente']))

# 5. Evaluar el modelo
y_pred_m = model_multiple.predict(X_test_m)
r2_simple = r2_score(y_test, model_simple.predict(X_test))
r2_multiple = r2_score(y_test_m, y_pred_m)

print(f"\nR虏 del modelo simple: {r2_simple:.4f}")
print(f"R虏 del modelo m煤ltiple: {r2_multiple:.4f}")
```

## 5\. Regresi贸n Lineal Polinomial 

驴Y si la relaci贸n no es una l铆nea recta perfecta? Podemos crear un modelo m谩s flexible a帽adiendo caracter铆sticas polinomiales (como $TV^2$). Esto nos permite ajustar una curva a los datos, 隆pero sigue siendo un modelo "lineal" en sus par谩metros\!

**Modelo a estimar (grado 2):** $Sales = \theta_0 + \theta_1 \times TV + \theta_2 \times TV^2$


```python
# Usaremos un Pipeline para simplificar los pasos de transformaci贸n y modelado
# Grado del polinomio
degree = 1

# 1. Crear el pipeline
# Paso 1: Generar caracter铆sticas polinomiales (1, TV, TV^2)
# Paso 2: Ajustar un modelo lineal a esas caracter铆sticas
pipeline = Pipeline([
    ('poly_features', PolynomialFeatures(degree=degree, include_bias=False)),
    ('lin_reg', LinearRegression())
])

# Usaremos las mismas variables X e y del modelo simple
pipeline.fit(X_train, y_train)

# 2. Evaluar el modelo
y_pred_poly = pipeline.predict(X_test)
r2_poly = r2_score(y_test, y_pred_poly)
print(f"R虏 del modelo simple (TV): {r2_simple:.4f}")
print(f"R虏 del modelo polinomial (Grado 2 en TV): {r2_poly:.4f}")

# 3. Visualizar el resultado
X_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_range_pred = pipeline.predict(X_range)

plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Datos Reales')
plt.plot(X_range, y_range_pred, color='purple', linewidth=2, label=f'Ajuste Polinomial (Grado {degree})')
plt.title('Regresi贸n Polinomial: Ventas vs. Gasto en TV')
plt.xlabel('Gasto en TV (miles de USD)')
plt.ylabel('Ventas (miles de unidades)')
plt.legend()
plt.grid(True)
plt.show()
```

### 锔 **Peligro: 隆Sobreajuste (Overfitting)\!**

驴Qu茅 pasa si usamos un grado de polinomio muy alto?


```python
# Ajustemos un modelo de grado 15
degree_overfit = 15
pipeline_overfit = Pipeline([
    ('poly_features', PolynomialFeatures(degree=degree_overfit, include_bias=False)),
    ('lin_reg', LinearRegression())
])
pipeline_overfit.fit(X_train, y_train)

# Visualizar el sobreajuste
y_range_overfit = pipeline_overfit.predict(X_range)
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue')
plt.plot(X_range, y_range_overfit, color='red', linewidth=2, label=f'Ajuste Polinomial (Grado {degree_overfit})')
plt.title('Ejemplo de Sobreajuste (Overfitting)')
plt.ylim(y.min()-5, y.max()+5) # Ajustar el eje y para ver la curva
plt.legend()
plt.show()
```

1. R: Al aumentar el grado del polinomio la precion del model se reduce
